```python
import numpy as np

# Define the matrix A
#Se debe aplicar el criterio de convergencia a mano
A = np.array([
    [4, -1, 0, 0],
    [-1, 4, -1, 0],
    [0, -1, 4, -1],
    [0, 0, -1, 4]
])

# Check if the matrix is diagonally dominant
is_diagonally_dominant = True
for i in range(A.shape[0]):
    diagonal_element = abs(A[i, i])
    off_diagonal_sum = np.sum(np.abs(A[i, :])) - diagonal_element
    if diagonal_element <= off_diagonal_sum:
        is_diagonally_dominant = False
        break

if is_diagonally_dominant:
    print('The matrix is diagonally dominant.')
else:
    print('The matrix is not diagonally dominant.')
```

```python
# Define the constants and initial values
b = np.array([1, 1, 1, 1]) 
x_init = np.array([0, 0, 0, 0])
epsilon = 1e-5
max_iterations = 100

# initialize variables
x_current = np.array([0.0, 0.0, 0.0, 0.0])  # Make sure to initialize as float
iteration_count = 0
converged = False
iterations_table = []

# Run the Gauss-Seidel loop again
while not converged and iteration_count < max_iterations:
    x_prev = np.copy(x_current)
    
    # Update x1, x2, x3, x4 using Gauss-Seidel method
    #Se despeja cada variable de la ecuacion de manera manual
    x_current[0] = round((1 + x_prev[1]) / 4, 5)
    x_current[1] = round((1 + x_current[0] + x_prev[2]) / 4, 5)
    x_current[2] = round((1 + x_current[1] + x_prev[3]) / 4, 5)
    x_current[3] = round((1 + x_current[2]) / 4, 5)
    
    # Debugging: Check if x_current is actually getting updated
    if np.all(x_current == x_prev):
        break
    
    # Store the current iteration values for the table
    iterations_table.append(np.copy(x_current))
    
    # Check for convergence
    if np.linalg.norm(x_current - x_prev, np.inf) < epsilon:
        converged = True
    
    iteration_count += 1

# Convert iterations_table to a NumPy array for better display
iterations_table = np.array(iterations_table)

iterations_table, iteration_count, x_current
```
