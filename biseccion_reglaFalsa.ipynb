```python
import math
import numpy as np
import pandas as pd
```

```python
def funcion(x):
    # Define tu función aquí
    return x**3 + math.log(math.sqrt(x)) + math.sin(x) - 2  # Ejemplo: x^2 - 4 = 0 tiene una raíz en x = 2

def metodo_biseccion(funcion, a, b, tolerancia, max_iteraciones):
    if funcion(a) * funcion(b) >= 0:
        print("La función no cumple con el requisito de cambio de signo en [a, b].")
        return None

    iteraciones = pd.DataFrame(columns=['iteracion', 'a', 'b', 'c', 'f(a)', 'f(b)', 'f(c)'])
    iteracion = 1
    while (b - a) / 2.0 > tolerancia:
        #agregar datos de iteracion a un dataframe
        iteraciones.loc[iteracion] = [iteracion, a, b, (a + b) / 2.0, funcion(a), funcion(b), funcion((a + b) / 2.0)]
        c = (a + b) / 2.0
        if funcion(c) == 0:
            break
        elif funcion(c) * funcion(a) < 0:
            b = c
        else:
            a = c

        if iteracion >= max_iteraciones:
            print(f"No se pudo encontrar una solución en {max_iteraciones} iteraciones.")
            return None

        iteracion += 1

    return (a + b) / 2.0 , iteraciones

def metodo_regla_falsa(funcion, a, b, tolerancia, max_iteraciones):
    if funcion(a) * funcion(b) >= 0:
        print("La función no cumple con el requisito de cambio de signo en [a, b].")
        return None

    iteracion = 1
    iteraciones = pd.DataFrame(columns=['iteracion', 'a', 'b', 'c', 'f(a)', 'f(b)', 'f(c)'])
    while abs(funcion(a)) > tolerancia and abs(funcion(b)) > tolerancia:
        #agregar datos de iteracion a un dataframe
        iteraciones.loc[iteracion] = [iteracion, a, b, (a * funcion(b) - b * funcion(a)) / (funcion(b) - funcion(a)), funcion(a), funcion(b), funcion((a * funcion(b) - b * funcion(a)) / (funcion(b) - funcion(a)))]
        c = (a * funcion(b) - b * funcion(a)) / (funcion(b) - funcion(a))
        if funcion(c) == 0:
            break
        elif funcion(c) * funcion(a) < 0:
            b = c
        else:
            a = c

        if iteracion >= max_iteraciones:
            print(f"No se pudo encontrar una solución en {max_iteraciones} iteraciones.")
            return None

        iteracion += 1

    return c, iteraciones

# Parámetros iniciales
a = 0.95
b = 1.1
tolerancia = 1e-5
max_iteraciones = 100
```

```python
# Llama al método de regla falsa
resultado = metodo_regla_falsa(funcion, a, b, tolerancia, max_iteraciones)


if resultado is not None:
    display(resultado[1])
    print(f"La raíz es: {resultado[0]:.5f}")
```
