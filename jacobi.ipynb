```python
import numpy as np

# Definir la matriz A y el vector b
A = np.array([[4, -1, 0, 0],
              [-1, 4, -1, 0],
              [0, -1, 4, -1],
              [0, 0, -1, 4]])

b = np.array([1, 1, 1, 1])

# Verificar si la matriz A es diagonalmente dominante
def is_diagonally_dominant(matrix):
    for i in range(matrix.shape[0]):
        diagonal_element = abs(matrix[i, i])
        row_sum = sum(abs(matrix[i, j]) for j in range(matrix.shape[1]) if j != i)
        if diagonal_element <= row_sum:
            return False
    return True

if is_diagonally_dominant(A):
    print("La matriz A es diagonalmente dominante")
else:
    print("La matriz A no es diagonalmente dominante")
```

```python
# Definir parámetros iniciales
x_init = np.array([0, 0, 0, 0])
epsilon = 1e-5
max_iterations = 1000

# Inicializar variables
x_init = np.array([0.0, 0.0, 0.0, 0.0])  # Inicializar con flotantes
x_current = x_init
iteration_count = 0
converged = False
table = []

# Método de Jacobi corregido
while not converged and iteration_count < max_iterations:
    x_next = np.zeros_like(x_current)
    for i in range(A.shape[0]):
        s = sum(A[i, j] * x_current[j] for j in range(A.shape[1]) if j != i)
        x_next[i] = (b[i] - s) / A[i, i]
    
    # Verificar convergencia
    if np.linalg.norm(x_next - x_current, np.inf) < epsilon:
        converged = True
    
    # Guardar la iteración actual para la tabla
    table.append((iteration_count, x_current.copy()))
    
    # Actualizar el vector y el contador de iteraciones
    x_current = x_next
    iteration_count += 1

# Guardar la última iteración para la tabla
table.append((iteration_count, x_current.copy()))

#x_current, iteration_count, table[:5]  # Mostrar el resultado y las primeras 5 filas de la tabla de iteraciones

x_rounded = np.round(x_current, 5)
x_rounded
```

```python
# Convertir la tabla de iteraciones a un formato más legible
import pandas as pd

# Crear un DataFrame para mostrar todas las iteraciones
df_iterations = pd.DataFrame(table, columns=['Iteración', 'Vector'])
df_iterations['x1'] = df_iterations['Vector'].apply(lambda x: round(x[0], 5))
df_iterations['x2'] = df_iterations['Vector'].apply(lambda x: round(x[1], 5))
df_iterations['x3'] = df_iterations['Vector'].apply(lambda x: round(x[2], 5))
df_iterations['x4'] = df_iterations['Vector'].apply(lambda x: round(x[3], 5))
df_iterations.drop('Vector', axis=1, inplace=True)

df_iterations
```
